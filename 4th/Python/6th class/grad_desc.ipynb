```python
# small lambda
import numpy as np
def E(w):
    return 3*(w-2)**2+5
def dE(w):
    return 6*(w-2)
eta=0.01;
w=np.zeros((11,1));
w[0]=5;
print('w(0)=%f' %(w[0]),'E(w(0))=%f' %(E(w[0])))
for i in range(0,10):
    w[i+1]=w[i]-eta*dE(w[i]);
    print('w(%d)=%f' %(i+1,w[i+1]),'E(w(%d))=%f'
    %(i+1,E(w[i+1])))
```

```python
# two vars
import numpy as np
from copy import copy
def E(w1,w2):
    return 3*(w1-2)**2+2*(w2-3)**2+3
def dE1(w1):
    return 6*(w1-2)
def dE2(w2):
    return 4*(w2-3)
w1=np.zeros((11,1)); w2=copy(w1);
eta=0.1; w1[0]=8; w2[0]=10;
print('(w1(0),w2(0))=(%f,%f)' %(w1[0],w2[0]),
      'E(w1(0),w2(0))= %f' %(E(w1[0],w2[0])))
for i in range(0,10):
    w1[i+1]=w1[i]-eta*dE1(w1[i])
    w2[i+1]=w2[i]-eta*dE2(w2[i])
    print('(w1(%d),w2(%d))=(%f,%f)' %(i+1,i+1,w1[i+1],w2[i+1]),
          'E(w1(%d),w2(%d))= %f' %(i+1,i+1,E(w1[i+1],w2[i+1]))) 
```

```python
# three vars
import numpy as np
from copy import copy
def E(w1,w2,w3):
    return 3*(w1-2)**2+2*(w2-3)**2+4*(w3-1)**2+3
def dE1(w1):
    return 6*(w1-2)
def dE2(w2):
    return 4*(w2-3)
def dE3(w3):
    return 8*(w3-1)
w1=np.zeros((11,1));w2=copy(w1);w3=copy(w1);
eta=0.1;w1[0]=8;w2[0]=10;w3[0]=5;
print('(w1(0),w2(0),w3(0))=(%f,%f,%f)' %(w1[0],w2[0],w3[0]))
print('E(w1(0),w2(0),w3(0))= %f' %(E(w1[0],w2[0],w3[0])))
for i in range(0,10):
    w1[i+1]=w1[i]-eta*dE1(w1[i])
    w2[i+1]=w2[i]-eta*dE2(w2[i])
    w3[i+1]=w3[i]-eta*dE3(w3[i])
    print('(w1(%d),w2(%d),w3(%d))=(%f,%f,%f)' 
          %(i+1,i+1,i+1,w1[i+1],w2[i+1],w3[i+1]))
    print('E(w(%d))= %f'%(i+1,E(w1[i+1],w2[i+1],w3[i+1])))
```

```python
# general N vars
import numpy as np
def E(w,N):
    return 0.5*np.sum((w-np.linspace(1,N,N)/N)**2)
def dE(w,t):
    return w-t 
N=10;w=0.5*np.ones((N+1,N));eta=0.7;
print('W(0)=(%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f)'
      %tuple(w[0,:]))
print('E(W(0))= %f' %(E(w[0,:],N)))
for i in range(0,N):
    w[i+1,:]=w[i,:]-eta*dE(w[i,:],np.linspace(1,N,N)/N)
    print('W(%d)='%(i+1),
          '(%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f,%.2f)'
          %tuple(w[i+1,:]))
    print('E(W(%d))= %f' %(i+1,E(w[i+1,:],N)))
```
